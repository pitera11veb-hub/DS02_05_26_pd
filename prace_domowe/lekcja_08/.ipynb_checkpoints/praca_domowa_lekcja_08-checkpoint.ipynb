```python
liczby = [-5, 3, -2, 8, 0, -1, 7, 4, -9, 6]

dodatnie = filter(lambda x: x > 0, liczby)

wynik = list(map(lambda x: x * 2, dodatnie))

print(wynik)
```

```python
sprzedaz = [
    (1, "elektronika", 45, 12500),
    (1, "odziez", 78, 4200),
    (2, "elektronika", 52, 15200),
    (2, "odziez", 65, 3800),
    (3, "elektronika", 38, 9800),
    (3, "odziez", 82, 5100),
    (4, "elektronika", 61, 18900),
    (4, "odziez", 71, 4500),
]

# 1. Tuple -> dict
def tuple_to_dict(rekord):
    return {
        "dzien": rekord[0],
        "kategoria": rekord[1],
        "liczba_transakcji": rekord[2],
        "wartosc": rekord[3]
    }

# 2. Dodanie średniej transakcji
def dodaj_srednia_transakcji(rekord):
    rekord["srednia_transakcja"] = (
        rekord["wartosc"] / rekord["liczba_transakcji"]
    )
    return rekord

# 3. Agregacja po kategorii
def agreguj_po_kategorii(dane):
    wynik = {}

    for rekord in dane:
        kat = rekord["kategoria"]
        wynik[kat] = wynik.get(kat, 0) + rekord["wartosc"]

    return wynik

# Przetwarzanie danych
dane = [tuple_to_dict(r) for r in sprzedaz]
dane = [dodaj_srednia_transakcji(r) for r in dane]

# Agregacja
agregacja = agreguj_po_kategorii(dane)

# Sortowanie malejąco po wartości
posortowane = sorted(
    agregacja.items(),
    key=lambda x: x[1],
    reverse=True
)

print("Agregacja:")
print(agregacja)

print("\nPosortowane:")
print(posortowane)
```
