# `Zadanie 9`

```python
cm = [
    [80, 10],  # TN, FP
    [5, 45]   # FN, TP
]

TN = cm[0][0]
FP = cm[0][1]
FN = cm[1][0]
TP = cm[1][1]

accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP)
recall = TP / (TP + FN)

print(f"accuracy: {accuracy:.2%}")
print(f"precision: {precision:.2%}")
print(f"recall: {recall:.2%}")
```
