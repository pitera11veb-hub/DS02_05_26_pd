```python
liczby = [-5, 3, -2, 8, 12, -1, 6, 0, 15, -3]

suma = 0
ile = 0

for liczba in liczby:
    if liczba > 0:
        suma += liczba
        ile += 1

print("Suma liczb dodatnich:", suma)
print("Liczba dodatnich wartości:", ile)
```

```python
suma = 0
liczba = 1

while liczba <= 50:
    suma += liczba
    liczba += 1

print("Suma:", suma)

wzor = 50 * (50 + 1) / 2
print("Wynik ze wzoru:", wzor)
```
